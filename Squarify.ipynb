```
import pandas as pd
mydata = pd.read_csv('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Visualization-Workshop/master/Datasets/water_usage.csv', index_col= 0)
```

```
labels = mydata['Usage'] + ' (' + mydata['Percentage'].astype('str') + '%)'
```

```
# Create figure
plt.figure(dpi=200)
# Create tree map
squarify.plot(sizes=mydata['Percentage'], label=labels, color=sns.light_palette('green', mydata.shape[0]))
plt.axis('off')
# Add title
plt.title('Water usage')
# Show plot
plt.show()
```
